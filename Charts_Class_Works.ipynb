### Subplots
### ax=plt.subplot(n,m,i) or plt.subplot(nmi)
### places ax at position i in n x m grid

### n = number of rows
### m = number of columns
### i = is the position you want your plot to be

```python
#plot 1
plt.figure(figsize=(10,8))
x=np.array
y=np.array
plt.subplot(2,2,1)
plt.plot(gas["Year"],gas['Canada'],'--oy',label='CANADA')
plt.legend()
plt.title('Gas prices of Canada')

#plot2
plt.subplot(2,2,2)
plt.plot(gas['Year'],gas['Italy'],'s:b',label='ITALY')
plt.legend()
plt.title('Gas price of Italy')



#plot3
plt.subplot(2,2,3)
plt.plot(gas['Year'],gas['France'],':ok',label='FRANCE')
plt.legend()
plt.title('Gas price of France')



#plot4
plt.subplot(2,2,4)
plt.plot(gas['Year'],gas['Japan'],'*-r',label='JAPAN')
plt.legend()
plt.title('Gas price of Japan')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,5))
labels=['Mon','Tue','Wed','Thu','Fri']
mobile_sales=[20,34,30,35,27]
pc_sales=[25,32,34,20,25]
x=np.arange(len(labels)) #the label location
width=0.35 #the width of the bars
a=plt.bar(x-width/2,mobile_sales,width,color='m', label='Mobile')
b=plt.bar(x+width/2,pc_sales,width,color='b', label='PC')
plt.ylabel('Total sales')
plt.title('Total sales by mobile and pc')
plt.xticks(x,labels)
plt.legend()
plt.bar_label(a)
plt.bar_label(b)
plt.savefig('groupedbar.png')
```

```python
#data for 2016 olympics
plt.figure(figsize=(10,7))
x=['USA','GBR','China','Russia','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,17,26,19,15]
bottom_b=[a+b for a,b in zip(gold,silver)]
a=plt.bar(x,gold,label='gold',color='r')
b=plt.bar(x,silver,bottom=gold,label='silver',color='y')
c=plt.bar(x,bronze,bottom=bottom_b,label='bronze',color='g')
plt.bar_label(a,label_type='center')
plt.bar_label(b,label_type='center')
plt.bar_label(c,label_type='center')
plt.legend()
plt.title('olympic medals by Countries')
plt.ylabel('Medals')
plt.xlabel('Countries')
plt.savefig('stackedbar.png')
plt.show()
```
