```python
import pandas as pd
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

```python
age=fifa.loc[fifa['Age']].count()
age
```

```python
old=fifa.loc[fifa['Age']>=30]
old=old['Age'].count()
old

#france=gas_prices.loc[df['France']>=5]
#france['France']
```

```python
normal=fifa.loc[fifa['Age']>=20]
normal=normal['Age'].count()
normal
```

```python
young=fifa.loc[fifa['Age']>=20]
young=young['Age'].count()
young
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,7))
plt.pie([old,normal,young],labels=['OLD','NORMAL','YOUNG'],colors=['#05fceb','#fcfc05', 'r'],
        autopct='% .0f %%', explode=[0.2,0.0,0.0])
plt.title('THE OLD, YOUNG AND NORMAL')
plt.legend()
plt.savefig('Old_normal_young.png')
plt.show()
```
